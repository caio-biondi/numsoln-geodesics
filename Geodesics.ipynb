## Geodesics

## Christoffel Symbols
$$\begin{align*}\Gamma_{ij} ^ u & = \frac{1}{2} \sum\limits_{\alpha = u, v} g ^ {u\alpha}\left( \frac{\partial g_{\alpha i}}{\partial p ^ j} + \frac{\partial g_{\alpha j}}{\partial p ^ i} - \frac{\partial g_{ij}}{\partial p ^ \alpha}\right) \\ \Gamma_{ij} ^ v & = \frac{1}{2} \sum\limits_{\alpha = u, v} g ^ {v\alpha}\left( \frac{\partial g_{\alpha i}}{\partial p ^ j} + \frac{\partial g_{\alpha j}}{\partial p ^ i} - \frac{\partial g_{ij}}{\partial p ^ \alpha}\right) \end{align*}$$


## Gauss Equations
$$\begin{alignat}{3} p_{uu} & = \Gamma_{uu} ^ u p_u && + \Gamma_{uu} ^ vp_v && + Le \\ p_{uv} & = \Gamma_{uv} ^ u p_u && + \Gamma_{uv} ^ v p_v && + M e \\ p_{vu} & = \Gamma_{vu} ^ u p_u && + \Gamma_{vu} ^ v p_v && + Me \\ p_{vv} & = \Gamma_{vv} ^ u p_u && + \Gamma_{vv} ^ v p_v && + Ne \end{alignat}$$

## Geodesic parametrization
$$c(t) = (c ^ 1(t), c ^ 2 (t))$$

## Geodesic equation

$$\begin{align*}  & \frac{d ^ 2 c ^ k}{dt ^ 2} + \sum\limits_{i, j = 1} ^ 2 \frac{dc ^ i}{dt} \frac{dc ^ j}{dt} \Gamma_{ij} ^ k = 0 \\ \implies & \frac{d ^ 2 c ^ k}{dt ^ 2} + \frac{dc ^ 1}{dt} \frac{dc ^ 1}{dt} \Gamma_{11} ^ k + \frac{dc ^ 1}{dt} \frac{dc ^ 2}{dt} \Gamma_{12} ^ k + \frac{dc ^ 2}{dt} \frac{dc ^ 1}{dt} \Gamma_{21} ^ k + \frac{dc ^ 2}{dt} \frac{dc ^ 2}{dt} \Gamma_{22} ^ k = 0, \quad k = 1, 2 \end{align*}$$

## Example (Unit sphere)
Regular parametrization
$$X(u, v) = \left( \cos v \sin u, \sin v \sin u, \cos u\right), (u, v) \in [0, \pi] \times [0, 2\pi).$$
Compute partial derivative
$$X_u = \left( \cos u \cos v, \sin v \cos u, - \sin u \right), \quad X_v = \left( -\sin v \sin u, \cos v \sin u, 0 \right).$$
So, our metric is given by
$$g = \begin{pmatrix} 1 & 0 \\ 0 & \sin ^ 2 u \end{pmatrix}.$$
Using the above formulas to compute the Christoffel Symbols we get
$$\Gamma_{ij} ^ u = \begin{pmatrix} 0 & 0 \\ 0 & - \sin u \cos u \end{pmatrix}, \quad \Gamma_{ij} ^ v = \begin{pmatrix} 0 & \cot u \\ \cot u & 0 \end{pmatrix}.$$
So the geodesic equations on a unit sphere are given by
$$\begin{cases} \frac{d ^ 2 c ^ 1}{d t ^ 2} + \left( \frac{dc ^ 2}{dt} \right) ^ 2 \Gamma_{11} ^ 1  & = 0 \\ \frac{d ^ 2 c ^ 2}{dt ^ 2} + 2 \frac{dc ^ 1}{dt} \frac{dc ^ 2}{dt} \Gamma_{12} ^ 2  & = 0 \\ c_0 ^ 1 = c ^ 1(0), \frac{dc ^ 1}{dt}\bigg\rvert_{0} = \frac{dc ^ 1}{dt}(0) \\ c_0 ^ 2 = c ^ 2(0), \frac{dc ^ 2}{dt}\bigg\rvert_{0} = \frac{dc ^ 2}{dt}(0). \end{cases}$$
To solve the geodesic initial value problem by a numerical method, we can write this system of two non-linear second other differential equaitons as a system of four first-order differential equaitons:
$$\begin{cases} \frac{d}{dt}[c ^ 1] & = \frac{d c ^ 1}{dt} \\ \frac{d}{d t} \left(\frac{d c ^ 1}{dt} \right) & = - \left( \frac{dc ^ 2}{dt} \right) ^ 2 \Gamma_{11} ^ 1 \\ \frac{d}{dt} [c ^ 2] & = \frac{d c ^ 2}{dt} \\ \frac{d}{dt}\left( \frac{d c ^ 2}{dt}\right) & = - 2 \frac{dc ^ 1}{dt} \frac{dc ^ 2}{dt} \Gamma_{12} ^ 2. \end{cases}$$
### Numerical Solution
Write
$$\frac{d}{dt} \begin{pmatrix} c ^ 1 \\ \frac{d c ^ 1}{dt} \\ c ^ 2 \\ \frac{d c ^ 2}{dt} \end{pmatrix} = \begin{pmatrix} \frac{d c ^ 1}{dt} \\ - \left( \frac{dc ^ 2}{dt} \right) ^ 2 \Gamma_{11} ^ 1 \\ \frac{d c ^ 2}{dt} \\ - 2 \frac{dc ^ 1}{dt} \frac{dc ^ 2}{dt} \Gamma_{12} ^ 2 \end{pmatrix} \\ $$
Let $c_m ^ 1 \approx c ^ 1 (\Delta t)$ and $c_m ^ 2 \approx c ^ 2(\Delta t)$. Now, we replace the continous derivative be forward difference approximation (forward Euler's method) to get
$$\frac{c_{m + 1} ^ 1 - c_m ^ 1}{\Delta t} = \frac{dc_m ^ 1}{dt} \implies c_{m + 1} ^ 1 = \frac{dc_m ^ 1}{dt}\Delta t + c_m ^ 1, \quad \frac{c_{m + 1} ^ 2 - c_m ^ 2}{\Delta t} = \frac{dc_m ^ 2}{dt} \implies c_{m + 1} ^ 2 = \frac{dc_m ^ 2}{dt}\Delta t + c_m ^ 2$$
and
$$\frac{\frac{dc_{m + 1} ^ 1}{dt} - \frac{dc_m ^ 1}{dt}}{\Delta t} = -\left( \frac{dc ^ 2}{dt} \right) ^ 2\Gamma_{22} ^ 1 \implies \frac{dc_{m + 1} ^ 1}{dt} = \frac{dc_m ^ 1}{dt} - \left( \frac{dc ^ 2}{dt} \right) ^ 2\Gamma_{22} ^ 1\Delta t, \quad \frac{\frac{dc_{m + 1} ^ 2}{dt} - \frac{dc_m ^ 2}{dt}}{\Delta t} = -2\left(\frac{dc ^ 1}{dt} \frac{dc ^ 2}{dt}\right) \Gamma_{12} ^ 2 \implies \frac{dc_{m + 1} ^ 2}{dt} = \frac{dc_m ^ 2}{dt} - 2\left( \frac{dc ^ 1}{dt}\frac{dc ^ 2}{dt} \right)\Gamma_{12} ^ 2\Delta t$$

## Example(Torus)
$$X(u, v) = \left((a \cos u + b)\cos v, (a \cos u + b)\sin v, a \sin u\right), (u, v) \in [0, 2\pi] \times [0, 2\pi]$$

```python
a = 1
b = 3

# Christoffel symbols (Torus)
def gamma_112(u: float, v: float) -> float:
    return -a * np.sin(v) / (b + a * np.cos(v))

def gamma_211(u: float, v: float) -> float:
    return (b + a * np.cos(v)) * np.sin(v) / a
    
# Time setup
dt = (1. / 2.) ** 8
tf = 15. # Final time
numsteps = np.ceil(tf / dt).astype(int)

# Suitable initial conditions
c10 = 1.
c20 = 1.
dc10 = 1.
dc20 = 1.

# Solution vectors
c1s = [] 
c2s = [] 

# Append initial conditions
c1s.append(c10)
c2s.append(c20)
dc1s = dc10
dc2s = dc20

# Torus
for i in range(1, numsteps):
    # Apply discretization (Fwd Euler)
    cm1 = dc1s * dt + c1s[i - 1]
    cm2 = dc2s * dt + c2s[i - 1]
    
    dcm1 = dc1s - 2 * (dc1s * dc2s) * gamma_112(c1s[i - 1], c2s[i - 1]) * dt
    dcm2 = dc2s - (dc1s) ** 2 * gamma_211(c1s[i - 1], c2s[i - 1]) * dt
    
    # Append newly computed values to solution
    c1s.append(cm1)
    c2s.append(cm2)
    dc1s = dcm1
    dc2s = dcm2
    
# Compute X(c(t))
xc = (a * np.cos(c2s) + b) * np.cos(c1s)
yc = (a * np.cos(c2s) + b) * np.sin(c1s)
zc = a * np.sin(c2s)

# Plotting
fig = plt.figure(dpi = 200)
ax = fig.add_subplot(projection='3d')

Nf = 128

# Set up domain
u = np.linspace(0, 2 * np.pi, Nf)
v = np.linspace(0, 2 * np.pi, Nf)
u, v = np.meshgrid(u, v) # Why this?

x = (a * np.cos(v) + b) * np.cos(u)
y = (a * np.cos(v) + b) * np.sin(u)
z = a * np.sin(v)

# Plot the surface
ax.plot_surface(x, y, z, color = 'b', alpha = 0.5)
# Plot the geodesic
ax.plot(xc, yc, zc, color = 'k')

# Set an equal aspect ratio
ax.set_aspect('equal')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.zaxis.labelpad= - 4# <- change the value here (IDK)

ax.set(title = r'Regular torus $c_0^1=${0: .2f}, $c_0^2=${1: .2f}, $\frac{{dc^1}}{{dt}}(0)=${2:.1f}, $\frac{{dc^2}}{{dt}}(0)=${3:.1f}'.format(c10, c20, dc10, dc20))
plt.show()
# fig.savefig('Torus-Geodesic3')
```

# Abstract Surfaces
## Poincaré's half plane
The **Poicaré's half-plane model** is the upper half plane denoted as $H = \{(x, y) : y > 0; x, y \in \mathbb{R}\}$ together with the metric
$$(ds) ^ 2 = \frac{(dx) ^ 2 + (dy) ^ 2}{y ^ 2}.$$
There matrices for the Christoffel symbols are:
$$\Gamma_{ij} ^ 1 = \begin{pmatrix} 0 & - \frac{1}{y} \\ - \frac{1}{y} & 0 \end{pmatrix}, \Gamma_{ij} ^ 2 = \begin{pmatrix} \frac{1}{y} & 0 \\ 0 & -\frac{1}{y} \end{pmatrix}$$

```python
def hypg(c10 : float, c20 : float, dc10: float, dc20: float) -> list:
    """
        Solve the geodesic equation for Poincaré's plane.
        
        c10, c20: x and y coordinates for initial condition
        dc10, dc20 tangent vector inital condition
    """
    
    # Christoffel symbols (Hyperbolic plane)
    def gamma_112(u: float, v: float) -> float:
        return -1. / v

    def gamma_211(u: float, v: float) -> float:
        return 1. / v

    def gamma_222(u: float, v: float) -> float:
        return -1. / v
    # Time setup
    dt = (1. / 2.) ** 4
    tf = 10. # Final time
    numsteps = np.ceil(tf / dt).astype(int)

    # Suitable init conditions (args)
    c10 = c10
    c20 = c20
    dc10 = dc10
    dc20 = dc20

    # Solution vectors
    c1s = [] 
    c2s = [] 

    # Append initial conditions
    c1s.append(c10)
    c2s.append(c20)

    dc1s = dc10
    dc2s = dc20

    for i in range(1, numsteps):
        # Apply discretization (Fwd Euler)
        cm1 = dc1s * dt + c1s[i - 1]
        cm2 = dc2s * dt + c2s[i - 1]

        dcm1 = dc1s - 2 * (dc1s * dc2s) * gamma_112(c1s[i - 1], c2s[i - 1]) * dt
        dcm2 = dc2s - (dc1s) ** 2 * gamma_211(c1s[i - 1], c2s[i - 1]) * dt - (dc2s) ** 2 * gamma_222(c1s[i - 1], c2s[i - 1]) * dt

        # Append newly computed values to solution
        c1s.append(cm1)
        c2s.append(cm2)
        dc1s = dcm1
        dc2s = dcm2
    
    return c1s, c2s # return soln vectors

# THIS NEEDS MORE WORK
c0_all = [(1., 2.), (0., 2.), (-1., 2.), (2., 1.), (1., 1.), (0., 1.), (-1., 1.), (-2., 1.)] # Some init conditions (Clever way?)

# Soln for all init conditions
c1s_all = [] 
c2s_all = []

n_init = len(c0_all) # Num of init conditions

for i in range(n_init):
    for j in {-1, 1}: # Directions for tangent at init condition
        c1s, c2s = hypg(c0_all[i][0], c0_all[i][1], j, 0.)
        c1s_all.append(c1s)
        c2s_all.append(c2s)
        
    if c0_all[i][1] == 1.: # Skip repeats
        c1s, c2s = hypg(c0_all[i][0], 2 ** (-6), 0., 1.) # Solns that blow up :) 
        c1s_all.append(c1s)
        c2s_all.append(c2s)
    
# Plotting
fig, axs = plt.subplots(1, 1)
axs.set_xlim(-3, 3)
axs.set_ylim(0, 2)
axs.set_title('Geodesics Poincaré\'s half plane')
axs.set_xlabel('$x$')
axs.set_ylabel('$y$')

num_soln = len(c1s_all) #Num of solns

for i in range(num_soln):
    axs.plot(c1s_all[i], c2s_all[i], 'xkcd:darkblue')
    
axs.set_aspect('equal')
fig.savefig('PoincarePlane')
```
